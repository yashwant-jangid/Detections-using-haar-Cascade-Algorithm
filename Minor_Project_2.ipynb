## **Detections using haar-Cascade Algorithm**

```python
import numpy as np
import cv2
from google.colab.patches import cv2_imshow
```

**1. Eye Detection**

```python
!wget https://raw.githubusercontent.com/opencv/opencv/master/data/haarcascades/haarcascade_eye.xml
```

```python
!wget https://img.freepik.com/free-photo/medium-shot-people-with-crossed-arms_23-2149008944.jpg?w=2000
```

```python
img=cv2.imread("/content/p1.jpg")
eye_model=cv2.CascadeClassifier("/content/haarcascade_eye.xml")
eyes=eye_model.detectMultiScale(img,1.2,20)
for (x,y,w,h) in eyes:
  cv2.rectangle(img,(x,y),(x+w,y+h),(0,255,0),2)
cv2_imshow(img)
```

Every Eye has been detected under these Scale Factor(1.2) and Min. neighbor(20)

**2. Face Detection**

We are applying the face detection on the same sample image.

```python
!wget https://raw.githubusercontent.com/opencv/opencv/master/data/haarcascades/haarcascade_frontalface_default.xml
```

```python
img=cv2.imread("/content/p1.jpg")
face_model=cv2.CascadeClassifier("/content/haarcascade_frontalface_default.xml")
faces=face_model.detectMultiScale(img,1.2,5)
for (x,y,w,h) in faces:
  cv2.rectangle(img,(x,y),(x+w,y+h),(0,0,255),4)
cv2_imshow(img)
```

We have detected all the faces with 1.2 Scale factor and 5 Min. neighbor.


**3. Smile Detection**

```python
!wget https://raw.githubusercontent.com/opencv/opencv/master/data/haarcascades/haarcascade_smile.xml
```

```python
img=cv2.imread("/content/p1.jpg")
smile_model=cv2.CascadeClassifier("/content/haarcascade_smile.xml")
smiles=smile_model.detectMultiScale(img,1.2,90)
for (x,y,w,h) in smiles:
  cv2.rectangle(img,(x,y),(x+w,y+h),(0,255,255),5)
cv2_imshow(img)
```

This is the best results where we can detect all the smiles but with 3 false detections.

**4.Eye Glasses Detection**

```python
p2=cv2.imread("/content/p2.jpg")
glass_model=cv2.CascadeClassifier("/content/haarcascade_eye_tree_eyeglasses.xml")
glasses=glass_model.detectMultiScale(p2,1.2,2)
for (x,y,w,h) in glasses:
  cv2.rectangle(p2,(x,y),(x+w,y+h),(0,255,255),3)
cv2_imshow(p2)
```
